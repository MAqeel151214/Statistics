## Introduction to Statistics

Statistics is the science of collecting, organizing, analyzing, interpreting, and presenting data.

Why statistics matters:
- Helps make decisions using data.
- Identifies trends and patterns.
- Supports scientific research.
- Forms the foundation of machine learning and AI.

Workflow:
Data Collection → Data Cleaning → Analysis → Interpretation → Decision Making


## Descriptive Statistics

Descriptive statistics summarize and describe data.

Examples:
- Average exam score
- Most common product purchased
- Distribution of customer ages

Goal:
Describe existing data without making predictions.


## Inferential Statistics

Inferential statistics uses a sample to make conclusions about a larger population.

Examples:
- Election polling
- Medical trials
- Market research

Common tools:
- Probability
- Sampling
- Confidence intervals
- Hypothesis testing


## Population vs Sample

Population:
Entire group of interest.

Sample:
Subset selected from the population.

Example:
Population = All university students.
Sample = 500 selected students.


## Parameter vs Statistic

Parameter:
Numerical characteristic of a population.

Statistic:
Numerical characteristic of a sample.

Examples:
Population Mean (μ)
Sample Mean (x̄)
Population Standard Deviation (σ)
Sample Standard Deviation (s)


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [10, 20, 30, 40, 50]
pd.Series(data)
```


# Measures of Central Tendency

Central tendency identifies the center of a dataset.



## Mean

Mean = Sum of values / Number of values

Advantages:
- Uses all observations

Disadvantages:
- Sensitive to outliers


```python

data = [10,20,30,40,50]
mean = np.mean(data)
print("Mean =", mean)
```


## Median

Middle value after sorting.

Advantages:
- Robust against outliers


```python

data = [10,20,30,40,50]
median = np.median(data)
print("Median =", median)
```


## Mode

Most frequently occurring value.


```python

from statistics import mode

data = [1,2,2,2,3,4,5]
print("Mode =", mode(data))
```


## Weighted Mean

Used when values have different importance.


```python

values = [80,90,95]
weights = [0.2,0.3,0.5]

weighted_mean = np.average(values, weights=weights)
print("Weighted Mean =", weighted_mean)
```


## Trimmed Mean

Removes extreme values before calculating the mean.

Useful when outliers exist.


```python

from scipy.stats import trim_mean

data = [1,2,3,4,5,100]
print(trim_mean(data, 0.1))
```


# Measures of Dispersion

Dispersion measures variability in data.



## Range

Range = Maximum − Minimum


```python

data = [10,20,30,40,50]
print(max(data)-min(data))
```


## Variance

Average squared distance from the mean.

Higher variance means greater spread.


```python

data = [10,20,30,40,50]
print("Variance =", np.var(data))
```


## Standard Deviation

Square root of variance.

Most commonly used measure of spread.


```python

data = [10,20,30,40,50]
print("Standard Deviation =", np.std(data))
```


## Coefficient of Variation (CV)

CV = (Standard Deviation / Mean) × 100

Useful for comparing variability across datasets.


```python

data = [10,20,30,40,50]

mean = np.mean(data)
std = np.std(data)

cv = (std/mean)*100
print("CV =", cv,"%")
```


# Data Types

## Qualitative (Categorical)

### Nominal
Categories without order.
Examples:
- Gender
- Country
- Color

### Ordinal
Categories with order.
Examples:
- Rankings
- Satisfaction Levels

## Quantitative (Numerical)

### Discrete
Countable values.
Examples:
- Number of students

### Continuous
Measured values.
Examples:
- Height
- Weight
- Temperature



# Frequency Distribution Table


```python

data = ["A","B","A","C","B","A"]

freq = pd.Series(data).value_counts()
print(freq)
```


# Bar Chart

Used for categorical variables.


```python

categories = ["A","B","C"]
values = [30,20,10]

plt.figure(figsize=(5,3))
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```


# Histogram

Used for numerical data distributions.


```python

data = np.random.normal(50,10,1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=20)
plt.title("Histogram")
plt.show()
```


# Bivariate Analysis

Study relationship between two variables.



## Contingency Table

Used for two categorical variables.


```python

df = pd.DataFrame({
    "Gender":["M","F","M","F","M"],
    "Result":["Pass","Pass","Fail","Pass","Fail"]
})

table = pd.crosstab(df["Gender"], df["Result"])
table
```


## Scatter Plot

Used for two numerical variables.

Helps identify:
- Correlation
- Trends
- Outliers


```python

x = np.arange(1,21)
y = x*2 + np.random.randint(-5,5,20)

plt.figure(figsize=(6,4))
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

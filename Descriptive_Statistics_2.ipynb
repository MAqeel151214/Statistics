```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```


# Quantiles, Quartiles, Deciles and Percentiles

Quantiles divide sorted data into equal parts.

- Quartiles -> 4 parts
- Quintiles -> 5 parts
- Deciles -> 10 parts
- Percentiles -> 100 parts


```python

data = [78,82,84,88,91,93,94,96,98,99]

print("Q1:", np.percentile(data,25))
print("Q2:", np.percentile(data,50))
print("Q3:", np.percentile(data,75))
```


# Percentile Rank

Shows what percentage of observations lie below a value.


```python

from scipy.stats import percentileofscore

score = 94
rank = percentileofscore(data, score)

print(rank)
```


# Five Number Summary

1. Minimum
2. Q1
3. Median
4. Q3
5. Maximum


```python

summary = {
    "Min": np.min(data),
    "Q1": np.percentile(data,25),
    "Median": np.percentile(data,50),
    "Q3": np.percentile(data,75),
    "Max": np.max(data)
}

summary
```


# Interquartile Range (IQR)

IQR = Q3 - Q1

Measures spread of middle 50% of data.


```python

Q1 = np.percentile(data,25)
Q3 = np.percentile(data,75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("IQR =", IQR)
print("Lower Bound =", lower)
print("Upper Bound =", upper)
```


# Boxplots

A boxplot visualizes:

- Minimum
- Q1
- Median
- Q3
- Maximum
- Outliers


```python

dataset = [10,12,13,15,16,18,20,22,25,100]

plt.figure(figsize=(6,4))
plt.boxplot(dataset)
plt.title("Boxplot Example")
plt.show()
```


# Side by Side Boxplots

Used to compare distributions.


```python

group_a = np.random.normal(70,10,100)
group_b = np.random.normal(80,12,100)

plt.figure(figsize=(6,4))
plt.boxplot([group_a, group_b], labels=["A","B"])
plt.show()
```


# Scatterplots

Used to visualize relationship between two numerical variables.


```python

x = np.arange(1,21)
y = x*3 + np.random.randint(-10,10,20)

plt.figure(figsize=(6,4))
plt.scatter(x,y)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```


# Covariance

Measures how two variables move together.


```python

experience = np.array([2,5,8,12,13])
salary = np.array([1,2,5,12,10])

np.cov(experience, salary)
```


# Correlation

Measures strength and direction of relationship.
Range = -1 to +1


```python

np.corrcoef(experience, salary)
```


# Correlation vs Causation

Correlation does NOT imply causation.

Two variables may be correlated because of a third hidden variable.
